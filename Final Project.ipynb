```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.metrics import r2_score

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('Altered_sleep_cycle_productivity.csv')
df.columns = df.columns.str.strip()

num_summary = df.select_dtypes(include='number').describe()
cat_summary = df.select_dtypes(include='object').describe()
print("Numerical Summary:\n", num_summary)
print("\nCategorical Summary:\n", cat_summary)

num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], color='red', bins=20)
    plt.title(f'Distribution of {col}')
    plt.show()

bivariate_pairs = [
    ('Total Sleep Hours', 'Productivity Score'),
    ('Exercise (mins/day)', 'Mood Score'),
    ('Caffeine Intake (mg)', 'Sleep Quality'),
    ('Stress Level', 'Productivity Score'),
    ('Screen Time Before Bed (mins)', 'Sleep Quality')
]
for x, y in bivariate_pairs:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=x, y=y, alpha=0.6, color='blue')
    plt.title(f'{x} vs {y}')
    plt.show()

plt.figure(figsize=(10, 6))
sns.pairplot(df[['Total Sleep Hours', 'Caffeine Intake (mg)', 'Stress Level', 'Productivity Score']], palette='Blues')
plt.suptitle("Multivariate Pairplot", y=1.02)
plt.show()

df['Sleep Hours Category'] = pd.cut(df['Total Sleep Hours'], bins=5, labels=['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])

plt.figure(figsize=(8, 6))
sns.barplot(x='Sleep Hours Category', y='Productivity Score', data=df, palette='Blues')
plt.title('Average Productivity Score by Sleep Hours Category')
plt.show()

print("\nFindings Summary:")
print("- Total Sleep Hours, Exercise, and low Stress Level are positively associated with higher Productivity Scores.")
print("- Caffeine Intake and Screen Time Before Bed tend to reduce Sleep Quality.")
print("- Mood Score and Productivity Score both correlate with lifestyle factors.")
print("- Strong linear patterns suggest suitability for regression models.")

print("\nSuggested Modeling Technique:")
print("Regression Model: Since 'Productivity Score' is a continuous variable influenced by numeric predictors, "
      "a multiple linear regression model is suitable.")

features = ['Total Sleep Hours', 'Exercise (mins/day)', 'Caffeine Intake (mg)',
            'Screen Time Before Bed (mins)', 'Stress Level']
X = df[features].dropna()
y = df.loc[X.index, 'Productivity Score']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print(f"\nRegression Model R² Score: {r2:.2f}")

X_scaled = StandardScaler().fit_transform(df[features].dropna())
kmeans = KMeans(n_clusters=3, random_state=42).fit(X_scaled)
df['Cluster'] = -1
df.loc[df[features].dropna().index, 'Cluster'] = kmeans.labels_
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Total Sleep Hours', y='Productivity Score', hue='Cluster', palette='Blues')
plt.title("K-Means Clustering (3 Groups)")
plt.show()
```

After seeing that correlation could be implimented, we performed a data analysis that begins with the dataset by removing unnecessary columns. It uses Pearson correlation to calculate the estimated linear relationships between numerical variables and visualizes them with a gradient-styled correlation matrix. Consequently, we applied linear regression with an 80/20 train-test split to predict three separate outcomes, Productivity Score, Mood Score, and Stress level.

```python
# Using Correlation Analysis to ses the relationships between columns
from google.colab import files
uploaded = files.upload()

df = pd.read_csv('Altered_sleep_cycle_productivity.csv')

# Removing spaces in columns' names
df.columns = df.columns.str.strip()
# Excluding columns that are not needed
columns_exclude = ['Date', 'Person_ID', 'Gender']
df = df.drop(columns= columns_exclude)
display(df)

# Implimenting Correlation Analysis
numerical_df = df.select_dtypes(include=['number'])
corr = numerical_df.corr(method = 'pearson')
corr.style.background_gradient(cmap = 'coolwarm').format(precision=2)


# Cross-Validation

# Libararies for Cross-Validation
from sklearn import model_selection
from sklearn import linear_model

#For Productivity
output = 'Productivity Score'
df = df.dropna()

# Test/Train
x= np.array(df.drop(columns = [output]))
y = df[output].dropna()

x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.2)

#Getting the percantage for cross-validation
model = linear_model.LinearRegression()
model.fit(x_train, y_train)

print(model.score(x_test, y_test))

#For Mood
output = 'Mood Score'
df = df.dropna()

# Test/Train
x= np.array(df.drop(columns = [output]))
y = df[output].dropna()

x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.2)

#Getting the percantage for cross-validation
model = linear_model.LinearRegression()
model.fit(x_train, y_train)

print(model.score(x_test, y_test))

#For Stress Levels
output = 'Stress Level'
df = df.dropna()

# Test/Train
x= np.array(df.drop(columns = [output]))
y = df[output].dropna()

x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.2)

#Getting the percantage for cross-validation
model = linear_model.LinearRegression()
model.fit(x_train, y_train)

print(model.score(x_test, y_test))
```
